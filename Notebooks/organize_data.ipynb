## Organizar colunas com as informações de interesse

```python
#discursos
discurso = []
d = 0
tam_discursos = len(discursos_raw)
while d < tam_discursos:
    discurso.append(discursos_raw[d]['Conteudo'])
    d += 1
    
print(len(discurso))
print(len(discursos_raw))

df = pd.DataFrame(discurso, columns=["Discurso"]) 

df
```

```python
#NomeParlamentar
nomeParlamentar = []
d = 0
tam_discursos = len(discursos_raw)
while d < tam_discursos:
    nomeParlamentar.append(discursos_raw[d]['IdentificacaoParlamentar']['NomeParlamentar'])
    d += 1
    
print(len(nomeParlamentar))
print(len(discursos_raw))
df['NomeParlamentar'] = nomeParlamentar
df
```

```python
#NomeCompletoParlamentar
nomeCompletoParlamentar = []
d = 0
while d < tam_discursos:
    nomeCompletoParlamentar.append(discursos_raw[d]['IdentificacaoParlamentar']['NomeCompletoParlamentar'])
    d += 1
    
print(len(nomeCompletoParlamentar))
print(len(discursos_raw))

df['NomeCompletoParlamentar'] = nomeCompletoParlamentar
df
```

```python
#SexoParlamentar
sexoParlamentar = []
d = 0
while d < tam_discursos:
    sexoParlamentar.append(discursos_raw[d]['IdentificacaoParlamentar']['SexoParlamentar'])
    d += 1
    
print(len(sexoParlamentar))
print(len(discursos_raw))

df['SexoParlamentar'] = sexoParlamentar
df
```

```python
#SiglaPartidoParlamentar
SiglaPartidoParlamentar = []
d = 0
while d < tam_discursos:
    if "SiglaPartidoParlamentar" in discursos_raw[d]['IdentificacaoParlamentar']:
        SiglaPartidoParlamentar.append(discursos_raw[d]['IdentificacaoParlamentar']['SiglaPartidoParlamentar'])
    else:
        SiglaPartidoParlamentar.append("nan")
    d += 1
    
print(len(SiglaPartidoParlamentar))
print(len(discursos_raw))

df['SiglaPartidoParlamentar'] = SiglaPartidoParlamentar
df
```

```python
#UFParlamentar
UFParlamentar = []
d = 0
while d < tam_discursos:
    if "UfParlamentar" in discursos_raw[d]['IdentificacaoParlamentar']:
        UFParlamentar.append(discursos_raw[d]['IdentificacaoParlamentar']['UfParlamentar'])
    else:
        UFParlamentar.append("nan")
    d += 1
    
print(len(UFParlamentar))
print(len(discursos_raw))

df['UfParlamentar'] = UFParlamentar
df
```

```python
#TextoResumo
TextoResumo = []
d = 0
while d < tam_discursos:
    if "TextoResumo" in discursos_raw[d]['IdentificacaoPronunciamento']:
        TextoResumo.append(discursos_raw[d]['IdentificacaoPronunciamento']['TextoResumo'])
    else:
        TextoResumo.append("nan")
    d += 1
    
print(len(TextoResumo))
print(len(discursos_raw))

df['TextoResumo'] = TextoResumo
df
```

```python
#DataPronunciamento
DataPronunciamento = []
d = 0
tam_discursos = len(discursos_raw)
while d < tam_discursos:
    DataPronunciamento.append(discursos_raw[d]['IdentificacaoPronunciamento']['DataPronunciamento'])
    d += 1
    
print(len(DataPronunciamento))
print(len(discursos_raw))

df['DataPronunciamento'] = DataPronunciamento
df
```

```python
colunas = ['Discurso', 'NomeParlamentar', 'NomeCompletoParlamentar',
       'SexoParlamentar', 'SiglaPartidoParlamentar', 'UfParlamentar',
       'TextoResumo', 'DataPronunciamento','presidente']

row = {}
for x in colunas:
    row[x] = "NaN"

df3 = pd.DataFrame(columns = colunas)  
dfoutros2 = pd.DataFrame(columns = colunas)

i=0
n = 0
while i < len(df2):
    k = 0
    sigla = str(df2['SiglaPartidoParlamentar'][i])
    uf = str(df2['UfParlamentar'][i])
    if sigla == 'nan':
        n += 1  

        for x in colunas:
            row[x] = "NaN"
        #para cada atributo, adicionar a ao dicionário
        for j in df2.loc[i]:
            row[colunas[k]] = j
            k += 1
        dfoutros2 = dfoutros2.append(row, ignore_index=True)
        
    elif  uf == 'nan':
        n += 1  
        for x in colunas:
            row[x] = "NaN"
        #para cada atributo, adicionar a ao dicionário
        for j in df2.loc[i]:
            row[colunas[k]] = j
            k += 1
        dfoutros2 = dfoutros2.append(row, ignore_index=True)
    
    else: 
        for x in colunas:
            row[x] = "NaN"
        #para cada atributo, adicionar a ao dicionário
        for j in df2.loc[i]:
            row[colunas[k]] = j
            k += 1
        df3 = df3.append(row, ignore_index=True)
        
    i += 1
    
print(n)
print(len(df2))
print(len(dfoutros2))
print(len(df3))
```
